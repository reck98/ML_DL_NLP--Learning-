```python
### While Loop -> It continues to execute until the condition is true 

i = 1
while(1):  ## 1 -> Always true 
    print(i)
    i = i + 1
    if i == 10:
        break


print("----------------------------------------------------------")

count = 1
while(count <= 10): 
    print(count)
    count = count + 1
```

```python
ele = 0

## Only one time executed 
while (ele % 2 == 0):
    print(ele)
    ele = ele + 1 
```

```python
## Break statements 

i = -4
while (1):
    print(i)
    i = i + 1
    if i == 1:
        break   ## Breaks the loop
```

```python
## Continues

for i in range(10):
    if (i % 2 == 0):
        continue
    print(i)

print("--------------------------------------------------")

for i in range(1, 10, 2):
    print(i)
```

```python
# Pass operation 

for i in range(5):
    if (i == 3):
        pass   ## JUst like a null operation 
    print(i)
```

```python
## Nested Loops 

for i in range(3):
    for j in range(2):
        print(f"i = {i}, and j = {j}")
    
```

```python
import math

def isPrime(n):
    if (n <= 1):
        return False

    for i in range(2, int(math.sqrt(n) + 1)):
        if (n % i == 0):
            return False

    return True


n = int(input("Enter a number: "))  
for i in range(1, n + 1):
    if isPrime(i):
        print(i)
```
