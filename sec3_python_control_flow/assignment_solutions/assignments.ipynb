```python
num = int(input("Enter a Number: "))

sum = 0

while(num != 0):
    sum = sum + (num % 10)
    num = num // 10

print(f"The Sum of digits is : {sum}")
```

```python
first = 0; second = 1

n = int(input("Enter the number of terms: "))

if ( n == 1):
    print(first)
elif (n == 2):
    print(second)
else:
    print(first)
    print(second)
    for i in range(2, n):
        next = first + second
        print(next)
        first = second
        second = next
```
