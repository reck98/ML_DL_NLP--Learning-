```python
## Probability 
        ## --> The likelihood of an event happening
        
        ## Example: Tossing a fair coin 
                    ## Outcomes can be {H or T}
                    ## P(H) = P(T) = 0.5
                    
        ## Example: Rolling a fair die 
                    ## Outcomes can be {1,2,3,4,5,6}
                    ## P(1) = P(2) = P(3) = P(4) = P(5) = P(6) = 1/6
                    
                    ## P_X(1) = 1/6 , Here X denotes the random Variable
```

## Mutually Exclusive Events

```python
## Mutually Exclusive Events 
        ## --> Events that cannot happen at the same time
        
        ## Example: In tossing a coin 
                ## We cannot get H or T at the same time
                    ## Thus, Event of getting H and Event of getting T are mutually exclusive events
        
        
'''
        P(H or T) = P(H) + P(T) 
                            {Additive Rule of Mutually exclusive events}
'''


## Example: Rolling a dice 
                ## Sample Space = {1,2,3,4,5,6}
        
'''
        P(1 or 5) = P(1) + P(5) = 1/6 + 1/6 = 1/3
'''
```

## Non Mutually Exclusive Events

```python
## Non Mutually Exclusive Events 
        ## --> Events that can happen at the same time
        
        ## Example: Taking a card from a deck of 52 cards
                    ## Getting a king card 
                            ## now this king can be of any suit
                            ## Thus, there are 4 events that can happen at the same time
        
        ## The probabilty of getting a king or hearts 
        
'''
        P(K or hearts) = P(K) + P(hearts) - P(K and hearts)
                            {Additive Rule of Non mutually exclusive events}
'''

## P(K or hearts) = 4 / 52 + 13 / 52 - (1 / 52)
## P(K or hearts) = 16 / 52
```

## Multiplication Rule

```python
## Independent Events 
        ## --> Events that do not depend on each other
        
        ## Example: Tossing a coin 
                ## Lets suppose i get H in first toss and then T
                ## P(H) = 1/2 , but this probabilty is not going to affect the probabilty of getting T in the second toss
                ## P(T) = 1/2


## Dependent Events 
        ## --> Events that depend on each other, means that their probability affects each other
        
        ## Example: Lets suppose we take out a king card from a deck of 52 cards and then a queen card from the same deck
                ## Now the probability of getting a queen card is going to increase as we are taking out a king card from the deck
                
                ## Thus, the queen card probabilty depends on the first event that is taking out a king card
                
                ## P(king) = 4 / 52
                # P(Queen) = 4 / 51
                
'''
        {Independent Events}
                P(H and T) = P(H) * P(T)
        
        {Dependent Events}
                P(K and Q) = P(K) * P(Q / K)  
                                          --> conditional probability
                                          --> Idea of Bayes Theorem
'''
```
