```python
import matplotlib.pyplot as plt
import math
```

```python
x = [1,2,3,4,5,6]
y = list(map(lambda x : x ** 2, x))
```

```python
x,y
```

```python
## Line plot 

plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Line Plot')
```

```python
## Customization 

plt.plot(x, y, color= 'red', linestyle= '--', marker = 'o', markersize = 5)
plt.grid()
```

```python
x = [1,2,3,4,5,6]
y = list(map(lambda x : x ** 2, x))
y1 = list(map(lambda x : math.log(x), x))
y2 = list(map(lambda x : math.sin(x), x))

y, y1, y2
```

```python
plt.figure(figsize=(11,5))

plt.subplot(1,3,1)
plt.plot(x, y, color = 'green')
plt.title("Squared Function ")

plt.subplot(1,3,2)
plt.plot(x, y1, color = 'red')
plt.title('Logarithmic Function')

plt.subplot(1,3,3)
plt.plot(x, y2, color = 'blue')
plt.title('Sin Function')

```

```python
# Pie Chart 

labels = ['A', 'B', 'C', 'D', 'E']
values = [10,20,30,15,25]
colors = ['gold', 'yellowgreen', 'skyblue', 'red', 'blue']
explode = (0.01,0,0.1,0,0)

plt.pie(values, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%')
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.scatterplot(df, x='total_bill', y='tip')
plt.title('Scatter Plot')
```

```python
sns.lineplot(df, x = 'size', y = 'total_bill')
```

```python
df.head()
```

```python
## CAtegorical Variables 
plt.figure(figsize=(13,5))
plt.subplot(1,2,1)
sns.barplot(df, x='sex', y='total_bill')

plt.subplot(1,2,2)
sns.barplot(df, x = 'day', y = 'total_bill')
```

```python
## Box Plot 

sns.boxplot(df, x = 'day', y = 'total_bill')
```

```python
## Violin Plot 

sns.violinplot(df, x = 'day', y = 'total_bill')
```

```python
sns.histplot(df['total_bill'], bins=10, kde=True)
```

```python
sns.kdeplot(df, x = 'total_bill', fill=True)
```

```python
## Pair Plot 

sns.pairplot(df)
```
