```python
import logging

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s | %(name)s | %(levelname)s | %(message)s", 
    force=True
)

logging.info("App started")
logging.error("Something failed")

logging.warning("Hey this is warning")
```

```python
logging.debug("hey this is a debug message")
logging.warning("Hey this is warning")
logging.error("Hey this is error")
logging.critical("Hey this is critical")
logging.info("Hey this is info")
```
