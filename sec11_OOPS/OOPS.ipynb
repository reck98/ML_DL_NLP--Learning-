```python
## Modelling a bank Account

class BankAccount:
    def __init__(self, owner):
        self.owner = owner
        self.currBalance = 0
    
    def deposit(self, amt):
        self.currBalance += amt
        print(f"The AMount deposited successfully")
    
    def withdraw(self, amt):
        if (self.currBalance >= amt):
            self.currBalance -= amt
            print(f"Amount withdrawn successfully")
        else:
            print("Not Enough Money")
    
    def details(self):
        print(f"Owners Name = {self.owner}")        
        print(f"Current Balance = {self.currBalance}")        
```

```python
myAccount = BankAccount("Aashish")
```

```python
myAccount.details()
```

```python
myAccount.deposit(25000)
```

```python
myAccount.withdraw(5000)
```

```python
myAccount.details()
```

### Inheritance

```python

## Parent Class
class Car:
    def __init__(self, windows, doors, engineType):
        self.windows = windows
        self.doors = doors
        self.engineType = engineType
    
    def drive(self):
        print(f"The person is driving {self.engineType} Car")
```

```python
c1 = Car(4,5,"Petrol")
c1.drive()
```

```python
## Child Class
class Tesla(Car):
    def __init__(self, windows, doors, engineType, isSelfDriving):
        super().__init__(windows, doors, engineType)
        self.isSelfDriving = isSelfDriving
    
    
    def selfDrive(self):
        print(f"Self Driving the car")
    
```

```python
T1 = Tesla(5,6,"Electric", True)
```

```python
T1.drive()
```

```python
## Mulitple Inheritance 

class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print(f"{self.name} is speaking")

class Pet:
    def __init__(self, owner):
        self.owner = owner


class Dog(Animal, Pet):
    def  __init__(self, name, owner, isGood):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)
        self.isGood = isGood

```

```python
class Shape:
    def area(self):
        return "The area of the figure"

class Rectangle(Shape):
    
    def __init__(self, b, l):
        self.b = b
        self.l = l
    
    def area(self):
        return f'{(self.b * self.l):.2f}'

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    
    def area(self):
        return (3.14 * self.r * self.r).__round__(2)
```

```python
c1 = Circle(4)
r1 = Rectangle(10,25)

def areaOfShape(shape):
    return f"The area is : {(shape.area())}"
```

```python
print(areaOfShape(c1))
print(areaOfShape(r1))
```

```python
## Getter and setter Method

class Person():
    def __init__(self, name, age, gender):
        self.__name = name ## Private Attributes 
        self.__age = age
        self.gender = gender
    
    def getName(self):  ## Get Name Method
        return self.__name
    
    def setName(self, name): ## Set Name Method 
        self.__name = name
    
    def setAge(self, age):
        if age > 0:
            self.age = age
        else:
            # raise AttributeError("Age should be greater than 0")
            print("Age should be greater than 0")
    
    def details(self):
        print(f"Name: {self.__name} \nAge: {self.__age} \nGender: {self.gender}")
```

```python
person = Person("Aashish", 22, "Male")
```

```python
person.setName("Aashish Kumar")
print(person.getName())
person.setAge(-22)
person.setAge(22)
person.details()
```

### Abstraction 

```python
from abc import ABC, abstractmethod

## Abstract Base Classs
class Vehicle(ABC):
    
    @abstractmethod
    def drive(self, name):
        self.name = name
        pass

class Car(Vehicle):
    
    def drive(self):
        print("The car is being drived ")
        
    def startEngine(self):
        print(f"Engine is started")
        

audi = Car()  
```

##### Opeartor OverLoading

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, other):
        return Vector(self.x * other.x, self.y * other.y)
    
    def __truediv__(self, other):
        return Vector(self.x / other.x, self.y / other.y)
    
    def __str__(self):
        return f"Vector({self.x}, {self.y})"
```

```python
v1 = Vector(1, 2)
v2 = Vector(3, 4)

print(v1 + v2)
print(v1 - v2)
print(v1 * v2)
print(v1 / v2)
```

### Custom Exception

```python
from datetime import datetime

class Error(Exception):
    pass

class DOBException(Error):
    pass



try:    
    year = int(input("Enter Your DOB"))
    currYear = datetime.now().year

    age = currYear - year
    if age <= 30 and age >= 20:
        print("You are Eligible")
    else:
        raise DOBException(f"DOB Exception")
except DOBException as err:
    print(f"Error Ocurred as {err} - You are Not Eligible")
except Exception as err:
    print(f"An error Occured as {err}")

```
