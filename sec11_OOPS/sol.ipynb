```python

class Address:
    def __init__(self, street, city, zipCode):
        self.street = street
        self.city = city
        self.zipCode = zipCode
    
    def __str__(self):
        return f"{self.street}, {self.city} - {self.zipCode}"
    
class Person:
    def __init__(self, name, age, add):
        self.name = name
        self.age = age
        self.add = add
    
    def details(self):
        print(f"Name : {self.name}")
        print(f"Age : {self.age}")
        print(f"Address : {self.add}")
    
    
```

```python
address = Address("One Street", "Dhanbad", 826004)

p1 = Person("Aashish", 22, address)
```

```python
p1.details()
```

```python
import math

class MathOperations:
    @staticmethod
    def squareRoot(num):
        return math.sqrt(num)


ans = MathOperations.squareRoot(16)
print(ans)
```

```python
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species
    
class Dog(Animal):
    def __init__(self, name, species, breed):
        super().__init__(name, species)
        self.breed = breed
        
    def __str__(self):
        return f"Name : {self.name}"

    def bark(self):
        print(f"The {self.name} is Barking")
    
  

    
d1 = Dog("Buddy", "kutta", "gali ka")
```

```python
print(d1.name)
print(d1.species)
print(d1.breed)
print(d1)
```

```python
d1.bark()
```

```python
class A:
    @classmethod
    def show(cls):
        print("From A is showing")    
        
class B(A):
    @classmethod
    def show(cls):
        print("From B is showing")    
        
class C(A):
    @classmethod
    def show(cls):
        print("From C is showing")    
        
```

```python
C.show()
A.show()
B.show()
```

```python
class D(C,B):  ## First One overrides other
    pass

```

```python
D.show()
```
