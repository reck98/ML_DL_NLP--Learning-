## Z Test 

```python
'''
        1. Must have population std dev 
        2. Sample size must be >= 30

'''

## Example:

populationMean = 168
populationStdDev = 3.9
sampleSize = 36
sampleMean = 169.5



## Given the confidence level is 95%

confidenceLevel = 0.95
alpha = 1 - confidenceLevel

## Null Hypothesis H_0 : Mean of the population is  168
## Alternate Hypothesis H_1 : Mean of the population is not  168

## Based on the confidence level, we need to draw the decision boundary



```

```python
## if the Z is greater or less than this decison boundary, Null is rejected
```

```python
##  Now Z score = (sampleMean - populationMean) / (populationStdDev / sqrt(sampleSize))


zScore = (sampleMean - populationMean) / (populationStdDev / math.sqrt(sampleSize))
print(zScore)
# 2.3076923076923075

## Since Z score is 2.3, we have to reject the Null Hypothesis
```

```python
from scipy.stats import norm


populationMean = 5
populationStdDev = 0.5

sampleSize = 40
sampleMean = 4.8


## H_0 = mean is 5
## H_1 = mean is less than 5
    
        ## Left tailed test


significanceLevel = 0.02
confidenceLevel = 1 - significanceLevel
print(f"Confidence Level: {confidenceLevel}")

zScore = (sampleMean - populationMean) / (populationStdDev / math.sqrt(sampleSize))
print(f"Z Score: {zScore}")

pvalue = norm.cdf(zScore)
print(f"pvalue: {pvalue}")

print(f"Area under the curve with the Zscore {zScore} is {pvalue}")


if pvalue < significanceLevel:
    print("Null Hypothesis is rejected")
else:
    print("Fail to reject the Null Hypothesis")
```

## Student t distribution

```python
## If we don't have the population std dev, then how can we perform any test?

## Student's t distribution

'''
            z = (sampleMean - populationMean) / (populationStdDev / sqrt(sampleSize))
            
            t = (sampleMean - populationMean) / (sampleStdDev / sqrt(sampleSize))
'''

## We will use t table 

'''
        Degree of freedom = sampleSize - 1
'''

```

```python
## Example: 
from scipy.stats import t

populationMean = 100
sampleSize = 30
sampleMean = 140
sampleStdDev = 20

tScore = (sampleMean - populationMean) / (sampleStdDev / math.sqrt(sampleSize))
print(f"tScore: {tScore}")

dof = sampleSize - 1
print(f"Degree of Freedom: {dof}")

##  H_0 = mean is 100
##  H_1 = mean is not 100

## Assume the cf level is 95% 

alpha = 0.05
confidenceLevel = 1 - alpha

pvalue = 2 * t.sf(abs(tScore), dof)
print(f"pvalue: {pvalue}")

if pvalue < alpha:
    print("Null Hypothesis is rejected")
else:
    print("Fail to reject the Null Hypothesis")
    
tCritical = t.ppf(confidenceLevel + alpha / 2, dof)
print(f"tCritical: {tCritical}")

if tScore > tCritical:
    print("Null Hypothesis is rejected")
else:
    print("Fail to reject the Null Hypothesis")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Given values
df = 29
alpha = 0.05
t_score = 10.96   # your computed t value

# Generate x values
x = np.linspace(-5, 5, 1000)
y = t.pdf(x, df)

# Critical t value
t_critical = t.ppf(1 - alpha/2, df)

# Plot t distribution
plt.figure()
plt.plot(x, y)

# Shade rejection regions
x_left = x[x <= -t_critical]
x_right = x[x >= t_critical]

plt.fill_between(x_left, t.pdf(x_left, df), alpha=0.3)
plt.fill_between(x_right, t.pdf(x_right, df), alpha=0.3)

# Draw vertical lines
plt.axvline(-t_critical, linestyle="--")
plt.axvline(t_critical, linestyle="--")
plt.axvline(t_score, linestyle=":")

# Labels
plt.title("Two-Tailed t-Test Rejection Region")
plt.xlabel("t value")
plt.ylabel("Density")
plt.text(t_score, 0.04, "Observed t")

plt.show()
```
