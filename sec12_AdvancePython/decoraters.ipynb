```python
## Closures

def mainWelcome():
    msg = "Hey This is main message"
    def subWelcome():
        print("Hey this is subWelcome")

    return subWelcome()
```

```python
mainWelcome()
```

```python
def mainWel(func, lst):
    def sub():
        print(func(lst))
    return sub()
```

```python
l = [1,2,3,4,5]
mainWel(len, l)
```

```python
### Decoraters 


def mainWelcome(func):
    msg = "Hey This is main message"
    def subWelcome():
        func()
        print("Hey this is subWelcome")

    return subWelcome()



@mainWelcome
def wel():
    print("Welcome to the advanced Series")
```

```python
## Decoraters with arguments 

def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(3)
def print_message(message):
    print(message)

print_message("Hello, World!")
```
