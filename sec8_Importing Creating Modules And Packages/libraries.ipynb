```python
import json


filePath = 'sample.json'
data = {
    "name" : "Aashish",
    "age" : 25,
    "job" : "Student",
    "hey" : "THere"
}

print(type(data))

jsonStr = json.dumps(data)
print(jsonStr)
print(type(jsonStr))

jsonJson = json.loads(jsonStr)
print(jsonJson)
print(type(jsonJson))

if not os.path.exists(filePath):
    with open(filePath, 'w') as f:
        pass
        

with open(filePath, 'w') as f:
    json.dump(data, f, indent=4)
    
    
        
```

```python
import csv

with open('sample.csv', mode='w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['name', 'age'])
    writer.writerow(['Aashish', 32])
    
with open('sample.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
