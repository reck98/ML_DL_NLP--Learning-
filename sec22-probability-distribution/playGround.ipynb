## Probability Density Function (pdf)

```python
## Distribution of continuous random variable


from scipy.stats import gaussian_kde


ages = np.random.normal(loc=40, scale=10, size=10000)
kde = gaussian_kde(ages)

x = np.linspace(ages.min() - 5, ages.max() + 5, 1000)
pdf = kde(x)

plt.figure(figsize=(8, 5))
plt.plot(x, pdf)
plt.xlabel("Ages")
plt.ylabel("Probability Density")
plt.title("PDF (Estimated using KDE)")
plt.show()
```

```python
## Two important things to observe

            ## 1. Probabily Density  (The function of plot)
            ## 2. Area under the curve
```

```python
## Potting the cumulative density for this plot

cdf = np.cumsum(pdf)
cdf = cdf / cdf[-1]

plt.figure(figsize=(8, 5))
plt.plot(x, cdf)
plt.xlabel("Ages")
plt.ylabel("Cumulative Probability")
plt.title("CDF (from KDE)")

xPoint = 40
yPoint = np.interp(xPoint, x, cdf)

plt.scatter(xPoint, yPoint, color='red')
plt.axvline(xPoint, color='red', linestyle='dashed')
plt.axhline(yPoint, color='red', linestyle='dashed')

plt.show()
```

## 1. Bernouli Distribution

```python
## Simplest Discrete Probability Distribution 

## It represents the probability of a random variable that has exactly two outcomes

                ## Success (p)
                ## Failure (1 - p)


## Used to model binary outcomes, yes/no dataset 



## Example;
        ## Whether a person is diabetic or not

        ## Outcomes = {Diabetic, Non-Diabetic}
        
        ## P(Diabetic) = 0.05
        ## P(Non-Diabetic) = 0.95


'''
                0 <= p <= 1
                k = {0, 1}
                            k = 1 --> Success
                            k = 0 --> Failure
'''


## PMF : 
        ## Lets say a company launched a phone 'A'
        
        ## Now, use = 40% (k = 1)
                ## Not use = 60% (k = 0)
                

x = np.array([0, 1])
y = np.array([0.4, 0.6])


plt.figure(figsize=(5, 3))
sns.barplot(x=x, y=y)
plt.xlabel("Outcomes")
plt.ylabel("Probabilities")
plt.title("Probabilty Mass Function")
```

```python
## This is the PMF constructed for the given example

'''
            PMF = ( p ^ k ) * ( (1 - p) ^ (1 - k) )
'''

## Mean of the Bernoulli Distribution 

    ## In order to calculate the mean of the distribution, we need to calculate the expected value of the random variable
    
    
'''
                E(x) = sigma( k = 0 to 1)  [   k * p(k)  ]   
                
                                                E(x) --> Expected Value
'''

## Example : 
            ## p = 0.6 
            ## E(x) = 0 * 0.4 + 1 * 0.6 = 0.6
            
## Expected Value = 0.6
        ## therefor the mean is 0.6, i.e. nothing but the value of p


## Median of the Bernouli Distribution 

'''
                Median = 0 if p < 0.5
                       = [0 , 1] if p = 0.5
                       = 1 if p > 0.5
                       
A better way to write it as

                Median = 0,  if  p < q
                        = 0.5 ,  if p = q
                        = 1,  if p > q
'''


## Mode for the bernouli distribution

'''
                if p > q, then mode = p
                else mode = q
'''


## Variance of the Bernouli Distribution

        ## Consider k = {0, 1},  P(X = 1) = p = 0.6 and P(X = 0) = q = 0.4

        ## now, var = ( 0 - 0.6)^2 * 0.4 + (1 - 0.6)^2 * 0.6
                    # = 0.24 (This is nothing but p * q)
            
            
'''
                sigma^2  = p * q
'''
    
    

```

## 2. Binomial Distribution


```python
##  Discrete Random Variable 

## Each outcome has a binary Value 

## The experiments are done in n trials 

## Basically, a series of Bernouli trials is considered as Binomial Distribution



## Example: Tossing a coin 10 times
                ## n = 10


## Notation: 
'''
                Denoted by -->    B(n, p)
                
                n = number of trials
                p = probability of success [0,1]


        Successes, k belongs to {0,1,2,3,4,5,6,7,8 ..... n}
        
'''


## PMF 
'''

            P(n, k, p) = ( n combitorial k ) * ( p ^ k ) * ( (1 - p) ^ (n - k) )
            
            n = number of trials
            k = number of successes
            p = probability of success

'''

## Exmple: 

## What is the probability of 5 success in 10 trials? where the probability of success is 0.6
p = 0.6
n = 10
k = 5

def nCr(n, r):
    return math.factorial(n) // math.factorial(r) // math.factorial(n - r)

y = nCr(n, k) * (p ** k) * ((1 - p) ** (n - k))

print(y)  ## --> PMF for k = 5


## Graph is 
        ## X - axis --> k (No. of success)
        ## Y - axis --> PMF
        

p = 0.6
n = 20

y = np.array([nCr(n, k) * (p ** k) * ((1 - p) ** (n - k)) for k in range(n + 1)])

print(y)

plt.figure(figsize=(5, 3))
sns.scatterplot(x=np.arange(n + 1), y=y)
plt.xlabel("Number of Successes")
plt.ylabel("Probabilities")
plt.title("Binomial Distribution")
plt.show()
```

```python
## Mean 
'''
            mean = n * p
'''

## Variance 
'''
        variance = n * p * q
'''

```

```python
## Example: Tossing a coin 
## What is the probability to get exactly three heads in 5 tosses?

## n = 5, p = 0.5, k = 3

ans = nCr(5, 3) * (0.5 ** 3) * ((1 - 0.50) ** ( 5 - 3))
ans
```

```python
## Example: Quality Control 

## Inspecting 10 times in a factory, where each item can be defective of 10% 

n = 10; p = 0.1
k = np.arange(0, 11)

def binomialDistribution(n, k, p):
    ans = nCr(n, k) * (p ** k) * ((1 - p) ** (n - k))
    return ans

y = np.array([binomialDistribution(n, k, p) for k in k])
yCDF = np.cumsum(y)

plt.figure(figsize=(5, 3))
sns.scatterplot(x = k, y = y)
sns.scatterplot(x = k, y = yCDF, color = "red")
plt.xlabel("Number of Defective")
plt.ylabel("Probabilities")
plt.title("Binomial Distribution")
plt.show()

y_2 = binomialDistribution(n, 2, p)
print(y_2)
```

## 3. Poission Distribution

```python
## Discrete Probability Distribution   ( PMF )
        ## Discrete Random Variables

## Describe the number of events occuring in a fixed interval of time 

    ## Example
            ## Number of people visiting hospital every hour 
            
            
numOfPeople = np.array([1,2,3,3,4,1])  ## Y - axis  (frequency)
time = np.array([12,1,2,3,4,5])  ## X - axis       (time)


plt.figure(figsize=(5, 3))
sns.barplot(x = time, y = numOfPeople)
plt.xlabel("Time")
plt.ylabel("Number of People")
plt.show()
```

```python
## One important parameter is lamda 


'''
            lamda = expected number of events occuring in the interval 
'''

## PMF 
'''
            PMF = P(X) =   ( e ^ (-lamda)  * (lamda ^ x)  ) / x!
            
            lamda = expected number of events occuring in the interval 
            x = random varaible 
            

'''

## Exmaple, for lamda = 3, what is the probability of one person visiting the hospital in 5th hour?

def poissonDistribution(lamda, x):
    return ((lamda ** x)  * math.exp(-lamda)) / math.factorial(x)

ans = poissonDistribution(lamda= 3, x = 5)
print(ans)
```

```python
## Construction of PMF: 

## X-Axis is k --> The number of occurences of an event
## Y-Axis is P(X = k) --> Probability of k occurences


lamda1 = 5  ## Avg number of accidents 
lamda2 = 8
lamda3 = 10

x = np.arange(0, 26)   ## Number of accidents per hour 

y1 = np.array([poissonDistribution(lamda1, k) for k in x])
yCDF1 = np.cumsum(y1)

y2 = np.array([poissonDistribution(lamda2, k) for k in x])
yCDF2 = np.cumsum(y2)

y3 = np.array([poissonDistribution(lamda3, k) for k in x])
yCDF3 = np.cumsum(y3)



plt.figure(figsize=(10, 6))

sns.scatterplot(x = x, y = y1)
sns.scatterplot(x = x, y = yCDF1)
sns.lineplot(x=x, y=yCDF1, linewidth=0.85, color = 'blue')
sns.lineplot(x=x, y=y1, linewidth=0.85, label=" = 5", color = 'blue')

sns.scatterplot(x = x, y = y2)
sns.scatterplot(x = x, y = yCDF2)
sns.lineplot(x=x, y=yCDF2, linewidth=0.85, color = 'green')
sns.lineplot(x=x, y=y2, linewidth=0.85, label=" = 8", color = 'green')


sns.scatterplot(x = x, y = y3)
sns.scatterplot(x = x, y = yCDF3)
sns.lineplot(x=x, y=yCDF3, linewidth=0.85,  color = 'red')
sns.lineplot(x=x, y=y3, linewidth=0.85, label=" = 10", color = 'red')

plt.legend()
plt.title("Poisson Distribution")
plt.xlabel("Number of accidents per hour")
plt.ylabel("Probabilities")
plt.show()

## Mean 

'''
            mean = lamda * t
                        --> lamda = expected number of events occuring in the interval
                        --> t = time interval
'''


## Variance 

'''
            variance = lamda * t
                        --> lamda = expected number of events occuring in the interval
                        --> t = time interval
'''
```

## 3. Gaussian / Normal Distribution

```python
## Continuous Probability Distribution   ( PDF )
        ## Continuous Random Variables
        
## The curve is a bell curve 

## The center is around the mean = median = mode

## Symmetric about the mean


```

```python
## Notation:

'''
        N(, ^2)
                : Mean of the data
                : std dev
                ^2 : variance ,  belongs to R , and > 0
            
'''

## Formula: 
```

## Empirical rule:

```python
## The empirical rule says that: 
    
'''
    ## Pr( ( - )  <=  x <= ( + ) ) ~= 68%
    ## Pr( ( - 2)  <=  x <= ( + 2) ) ~= 95%
    ## Pr( ( - 3)  <=  x <= ( + 3) ) ~= 99.7%
'''
```

```python
x = np.array([1,2,3,4,5]) ## random variable


def gaussian(x, xMean, xStdDev):
    expVal = (math.pow((x-xMean), 2)) / (2 * math.pow(xStdDev, 2)) 
    val = pow((2 * math.pi), 0.5) * xStdDev * math.exp(expVal) 
    return 1 / val

xMean = np.mean(x)

xStdDev = np.std(x)


pdf = np.array([gaussian(val, xMean, xStdDev) for val in x])
cdf = np.cumsum(pdf)
print(xMean)
print(xStdDev)


plt.figure(figsize=(5, 3))

plt.subplot(1, 2, 1)
sns.lineplot(x=x, y=pdf)
plt.title("Gaussian Plot")
plt.xlabel("Random Variable")
plt.ylabel("Probability")


plt.subplot(1, 2, 2)
sns.lineplot(x = x, y = cdf, color = 'red')
plt.xlabel("Random Variable")
plt.ylabel("Probability")
```

## 4. Uniform Distribution

```python
## There can be two types of uniform distributions

    ## 1. Continuous
    ## 2. Discrete


## Continuous:


## The entire distribution is between a and b 

        ## a --> The minimum Value
        ## b --> The maximum value
        

'''
     Notation: 
                U(a, b)
                        a --> minimum value
                        b --> maximum value
                    -inf <= a <= b <= +inf
'''


'''

    Formula: 
    
            pdf = 1 / (b - a)   if a <= x <= b
                = 0             otherwise
            
            cdf = 0          if x < a
                = ( x - a ) / (b - a)   if a <= x <= b
                = 1             if x > b
            
          
          
    Mean: 
            mean = (a + b) / 2
    
    
    Median: 
            median = (a + b) / 2
            

    Variance: 

            variance = (b - a)^2  / 12
    

'''
```

```python
## Example:

    ## The number of candies sold at a shop is uniformly distributed between 10 and 40
    
    ## Q1 -> What is the probability of daily sales between 15 and 30
    
    
    ## soln:

a = 10
b = 40


def pdf(x):
    if a <= x <= b:
        return 1 / (b - a)
    else:
        return 0


def cdf(x):
    if x < a:
        return 0
    elif a <= x <= b:
        return (x - a) / (b - a)
    else:
        return 1
    
    
x = np.arange(0, b + (abs(a-0)), 1)
y = np.array([pdf(val) for val in x])


ans = cdf(30) - cdf(15)
print(ans)


plt.plot(x, y)
```

```python
## Q2. what is the probabilty of sales greater than 20?

ans = cdf(max(b, 20)) - cdf(20)
print(ans)
```

## 5. Log Normal Distribution

```python
## Right skewed distribution

## if a random variable X follows a log normal distrbution with  = 0 , 

## then we can calculate Y = ln(X) , such that Y follows a normal distribution 

## Now equivalently, 

'''
        ## X = exp(Y)
                X is the log normal distribution 
                Y is the normal distribution

'''

'''
Note: To check if a plot is normally distributed or not , 

        WE use QQ - PLOT 
'''
def gaussian(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2*np.pi))) * np.exp(-(x-mu)**2 / (2*sigma**2))

def lognormal(x, mu, sigma):
    return (1 / (x * sigma * np.sqrt(2*np.pi))) * np.exp(-(np.log(x)-mu)**2 / (2*sigma**2))

start, stop, size = -5, 6, 100
mean, stdDev = 0, 1


y = np.linspace(start, stop, size)
yNormal = gaussian(y, mean, stdDev)
yNormalCDF = np.cumsum(yNormal)

indxNormalMax = np.argmax(yNormal)

yForMaxValue = y[indxNormalMax]
yNormalMaxValue = yNormal[indxNormalMax]


x = np.array([math.exp(val) for val in y])  
xLogNormal = lognormal(x, mean, stdDev)
xLogNormalCDF = np.cumsum(xLogNormal)
xLogNormalMaxValue = np.max(xLogNormal)

plt.figure(figsize=(15,8))

#############################################################################################################
plt.subplot(2, 2, 1)
sns.lineplot(x=y, y=yNormal, label="Normal Gaussian")
plt.scatter(
    x=yForMaxValue, 
    y = yNormalMaxValue, 
    color = "black", 
    label = "Max value", 
    zorder = 5, 
    s=25
)

plt.vlines(x = yForMaxValue, ymin= 0, ymax=yNormalMaxValue, linestyles = (0, (6,6)), color = "black")


plt.hlines(
    y = yNormalMaxValue,
    xmin = -5, 
    xmax = yForMaxValue, 
    linestyles=(0,(6,6)), 
    color = "black"
)
plt.annotate(
    f"Max Value is {yNormalMaxValue.__round__(4)}",
    xy=(yForMaxValue, yNormalMaxValue),
    xytext=(yForMaxValue + 0.8, yNormalMaxValue + 0.05),
    arrowprops=dict(arrowstyle="->"),
)

plt.legend(
    loc = "upper right"
)

#############################################################################################################

plt.subplot(2, 2, 2)
sns.lineplot(x=y, y=yNormalCDF, label="CDF of Normal Gaussian", color = "darkgreen")

#############################################################################################################

plt.subplot(2, 2, 3)
sns.lineplot(x=x, y=xLogNormal, label="Log-Normal", color = "red")
plt.xscale("log")

#############################################################################################################


plt.subplot(2, 2, 4)
sns.lineplot(x = x, y = xLogNormalCDF, label = "CDF of Log Normal", color = "green")
plt.xscale("log")


#############################################################################################################

print(yNormalMaxValue)
print(xLogNormalMaxValue)
```
