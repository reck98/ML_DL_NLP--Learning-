```python
import sqlite3

conn = sqlite3.connect('./test.db')
conn.row_factory = sqlite3.Row
```

```python
cur = conn.cursor()
```

```python
cur.execute('''
            CREATE TABLE IF NOT EXISTS employees(
                id INTEGER PRIMARY KEY,
                name text NOT NULL,
                age INTEGER, 
                department text
            )
            ''')

conn.commit()
```

```python
cur.execute('''
            SELECT name FROM SQLITE_MASTER WHERE type = 'table'
            ''')

cur.fetchall()
```

```python
cur.execute('''
            SELECT * FROM employees 
            ''')
rows = cur.fetchall()

for row in rows:
    print(dict(row))
```

```python
cur.execute('''
            PRAGMA TABLE_INFO (employees)
            ''')
cur.fetchall()
```

```python
def employeeInfoAgeGreaterThan(age):
    cursor.execute('''
                   SELECT * FROM employees
                   WHERE age > ?
                   ''', (age, ))
    
    rows = cursor.fetchall()
    
    if not rows:
        print("No Data Found")
        return []
    
    return rows

def employeeInfoAgeGreaterThan_SHOW(age):
    cursor.execute('''
                   SELECT * FROM employees
                   WHERE age > ?
                   ''', (age, ))
    
    rows = cursor.fetchall()
    
    if not rows:
        print("No Data Found")
        
    for row in rows:
        print(row)
    
    return
    
```
