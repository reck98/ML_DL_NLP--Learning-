```python
def transposeMatrix(matrix):
    transposed = [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
    
    return transposed

matrix = [
    [4, 5, 6],
    [1, 2, 3],
    [8, 5, 9]
]

tran = transposeMatrix(matrix)

for rows in tran:
    print(rows)
```

```python
lst1 = [1,5,6,7,8,9,5]

def revList(lst : list):
    lst2 = lst.copy()
    lst2.reverse()
    
    return lst2

lst2 = revList(lst1)

print(lst2)

```
