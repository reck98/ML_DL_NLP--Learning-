```python
studentDetails = [
    {
        "name" : "Aashish",
        "age" : 32,
        "grade" : "A"
    },
    {
        "name" : "Aarunee",
        "age" : 12,
        "grade" : "A+"
    },
    {
        "name" : "Aditya",
        "age" : 42,
        "grade" : "B"
    }
]

for student in studentDetails:
    for keys in student:
        print(f"{keys} : {student[keys]}")
    print("------------------")
```

```python
studentDetails = {
    "123" : {
        "name" : "Aashish",
        "age" : 25
   },
    
    "456" : {
        "name" : "Aarunee",
        "age" : 35
   }
}

for idNum in studentDetails:
    print(f"idNum is {idNum}")
    for key in studentDetails[idNum]:
        print(f"{key} : {studentDetails[idNum][key]}")
    print("-----------------")
    
```

```python
### Dictonary Comprehension

d = {i : i ** 2 for i in range(11)}

for key in d:
    print(f"{key} : {d[key]}")
```

```python
evensd = {key : value for key, value in d.items() if value % 2 == 0}
print(evensd)
```

```python
nums = [2,5,6,6,5,4,6,2,6,5,6,1,3,6,1,3,5,2,2,2]

d = {}

for num in nums:
    if num in d:
        d[num] += 1
    else:
        d[num] = 1

print(d)
```
