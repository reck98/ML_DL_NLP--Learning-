```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

for fruit in fruits:
    print(fruit)

for i in range(len(fruits)):
    print(str(i + 1) + ". " + fruits[i])

for i in range(-1, 0 - len(fruits) - 1, -1):
    print(str(i) + ". " + fruits[i])
```

```python
## Slicing 
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
i  = 4
print(1, fruits[1:3])
print(2, fruits[1:])
print(3, fruits[:3])
print(i, fruits[:])
i = i + 1
print(i, fruits[-3:-1])
i = i + 1
print(i, fruits[::-1])
i = i + 1
print(i, fruits[::2])
i = i + 1
print(i, fruits[1::2])
i = i + 1
```

```python
## Modifying fruits 

fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

fruits[1] = "watermelon"
print(fruits)
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

fruits.sort()
print(fruits)

fruits.sort(reverse=True)
print(fruits)

fruits.reverse()
print(fruits)

fruits.clear()
print(fruits)
```

```python
## Iterate with index

fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

for i, fruit in enumerate(fruits):  ## Returns index and value 
    print(f"Fruit present at indesx {i} is {fruit}")
```

```python
## List Comprehension 

lst = []

for i in range(10):
    lst.append(i ** 2)

print(lst)
```

```python
lst = [i ** 2 for i in range(10)]

print(lst)
```

```python
evenNumbers = [i for i in range(10) if i % 2 == 0]

print(evenNumbers)
```

```python
lst1 = [1,2,3,4]; lst2 = ['a', 'b', 'c', 'd']

pairs = [[i, j] for i in lst1 for j in lst2]

print(pairs)
```

```python
words = ["apple", "banana", "cherry", "kiwi", "mango"]

lengths = [len(word) for word in words]

lengths.sort()
print(lengths)
```
