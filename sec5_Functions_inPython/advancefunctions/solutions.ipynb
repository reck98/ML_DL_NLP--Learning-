```python
def fibo(n):
    if n <= 1:
        return n
    
    return fibo(n-1) + fibo(n - 2)

for i in range(11):
    print(fibo(i), end=" ")
```

```python
def onlyInteger(**kwargs):
    d = {}
    for key, value in kwargs.items():
        if isinstance(value, int):
            d[key] = value
    return d


d = onlyInteger(name=25, key=24, hey="125", heynot="knjhv")
print(d)
```

```python
def applyFunction(iterable, func):
    newList = list(map(func, iterable))
    
    return newList

nums = [1,2,6]

def sqaure(n):
    return n * n

print(applyFunction(nums, sqaure))
```

```python
def retSquare():
    def sqaure(a):
        return a * a
    return sqaure

sq = retSquare()

print(sq(3))
```
