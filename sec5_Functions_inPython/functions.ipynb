```python
## Positonal and keyword arguments 

def addNumbers(*args):  #positional
    res = 0
    for i in args:
        res = res + i

    return f"The total sum is {res}"
```

```python
res = addNumbers(25,25,25,25,45)
print(type(res))
print(res)
```

```python
# Keyword Argument 


def printDetails(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")


printDetails(name = "Aashish", age = 25, job = "Student")
```

```python
def printDetails(*args, **kwargs):
    for val in args:
        print(val)
    for key, value in kwargs.items():
        print(f"{key} : {value}")
```

```python
printDetails(25,25,25,"HEy", name = "Aashish", age = "255")
```

```python
## Return Statement

def mul(*args):
    res = 1
    for i in args:
        res = res * i
    
    return res
```

```python
print(mul(25,25,25,25,2,5,2,5,25,2,5,2))
```

##### Temp

```python
def tempConverter(temp, fromUnit : str = 'F', toUnit : str = 'C'):
    if fromUnit == toUnit:
        return temp
    if fromUnit == 'F' and toUnit == 'C':
        return ((temp - 32) * 5 / 9).__round__(2)
    elif fromUnit == 'C' and toUnit == 'F':
        return (temp * 1.8  + 32).__round__(2)
    else:
        return "Invaild Arguments"
        
     
    
```

```python
res = tempConverter(25,'F','C')
print(res)

temp = int(input("Enter Value: "))
fromUnit = str(input("Enter from: ")).strip().upper()
toUnit = str(input("Enter to: ")).strip().upper()

print(fromUnit)
print(toUnit)
print(tempConverter(temp, fromUnit, toUnit))
```

##### Password Strength


```python
def isStrongPassword(password):
    
    if (len(password) < 8) or (not any(char.isdigit() for char in password)) or (not any(char.islower() for char in password)) or (not any(char.isupper() for char in password)):
        return False
    if not any(char in "!@#$^&*()?" for char in password):
        return False
    
    return True

print(isStrongPassword("dhjsYaaa2?????T"))
    
    
```

```python
def calCost(cart):
    cost = 0
    for item in cart:
        total = item['price'] * item['quantity']
        cost += total
    
    return cost


cart = [
    {
        "name" : "Apple",
        "price" : 25,
        "quantity" : 5
    },
    {
        "name" : "Banana",
        "price" : 15,
        "quantity" : 2
    },
    {
        "name" : "Kiwi",
        "price" : 14,
        "quantity" : 10
    },
    {
        "name" : "Orange",
        "price" : 25,
        "quantity" : 4
    }
]

print(f"The Total Cost is : {calCost(cart)}")
```

```python
def isPallindrome(S : str):
    S = S.lower().strip()
    return (S[::-1] == S)


print(isPallindrome("a man a plan a canal Panama"))
```

```python
def factRecur(n):
    if ( n <= 1):
        return 1
    
    ans = n * factRecur(n-1)
    
    return ans

print(factRecur(5))
```

```python
def countWords(filePath):
    wordCount = {}
    with open(filePath, 'r') as f:
        for line in f:
            words = line.split()
            for word in words:
                word = word.lower().strip("./,?;:\"'")
                if word in wordCount:
                    wordCount[word] += 1
                else:
                    wordCount[word] = 1
    
    return wordCount   

filePath = "E:\Development\ML_DL_NLP\sec5_Functions_inPython\pyproject.toml"

print(countWords(filePath))
     
```
