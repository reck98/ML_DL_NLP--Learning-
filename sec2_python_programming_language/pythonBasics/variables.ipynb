```python
# Declaring and Assigning Variables 


a = 25   # integer
name = "Aashish" # string
height = 5.5   # float
isStudent = False; isEngineer = True # boolean


```

```python
print("a =", a)
print("name =", name)
print("height =", height)
print("isStudent =", isStudent)
print("isEngineer =", isEngineer)
```

```python
## Naming Convenetions

# 1. Variable name must be descriptive 
# 2. must start with a letter or an underscore and can contain numbers and underscores
# 3. Variable names are case-sensitive

# Examples

# name-1 = "Aashish" # Invalid
name_1 = "Aashish" # Valid
Name_1 = "Aashish2" # Valid
name1 = "Aashish" # Valid
NAME_1 = "Aashish" # Valid

name = "Person 1"
Name = "Person 2"

print(name == Name)  # False
```

```python
# Python is Dynamically Typed -> The type of variable is determined at runtime

age = 25 # int 
height = 5.5 # float
name = "Aashish" # string

print(type(age))
print(type(height))
print(type(name))
```

```python
## Type Conversion

age = 25
print(type(age))
# age = float(age)  ## Converting int to float
# print(type(age))

ageStr = str(age)
print(type(ageStr))

age = int(ageStr)  ## Converting string to int (but it will throw an error if the string is not a valid int)
print(type(age))


var = 25
varFloat = float(var)
varStr = str(varFloat)
# varInt = int(varStr) # Error in varInt as the varStr is not a valid int (it is "25.0")


print(var, varFloat, varStr)  
```

```python
# Dynamic Typing

var = 10
print(var, type(var))

var = 2.5
print(var, type(var))

var = "Hellow"
print(var, type(var))
```

```python
## input 

age = input("Enter Your Age: ")

print(age, type(age))  # by default it is a string

age = int(input("Enter Your Age: "))

print(age, type(age))
```

```python
## Basic Calculator 

num1 = int(input("Enter First Number: "))
num2 = int(input("Enter Second Number: "))

add = num1 + num2
sub = num1 - num2
mul = num1 * num2
div = num1 / num2

print("Addition: ", add)
print("Subtraction: ", sub)
print("Multiplication: ", mul)
print("Division: ", div)
```
