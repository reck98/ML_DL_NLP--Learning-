```python
## Arithmetic Operators

a = 25
b = 6

add = a + b # add
sub = a - b # subtract
mul = a * b # multiply
div = a / b # divide
floorDiv = a // b # floor division
mod = a % b # modulus
exp = a ** b # exponent

print(add)
print(sub)
print(mul)
print(div)
print(floorDiv)
print(mod)
print(exp)
```

```python
## Comparison Operators

a = 25
b = 45

print(a == b) # equal to
print(a != b) # not equal to
print(a > b) # greater than
print(a < b) # less than
print(a >= b) # greater than or equal to
print(a <= b) # less than or equal to
```

```python
str1 = "Aashish"
str2 = "aashish"

print(str1 == str2) 
```

```python
## Logical Operators

a = True
b = False

print(a and b) # and
print(a or b) # or
print(not a) # not
xorRes = a ^ b # xor
print(xorRes)

xorRes2 = (a and (not b)) or ((not a) and b)
print(xorRes2)
```

```python
## Calculator Version 2 

num1 = float(input("Enter 1st number: "))
num2 = float(input("Enter 2nd number: "))

add = num1 + num2
sub = num1 - num2
mul = num1 * num2
div = num1 / num2
floorDiv = num1 // num2
mod = num1 % num2
exponent = num1 ** num2


print(f"Addition: {add}")
print(f"Subtraction: {sub}")
print(f"Multiplication: {mul}")
print(f"Division: {div.__round__(2)}")
print(f"Floor Division: {floorDiv}")
print(f"Modulus: {mod}")
print(f"Exponent: {exponent}")
```
